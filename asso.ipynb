```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from datetime import datetime
import seaborn as sns
```

```python
URL = 'https://www.data.gouv.fr/fr/datasets/r/403c9108-dd0a-443d-9ed5-bca5f056fa8f'
df = pd.read_csv(URL, error_bad_lines=False, sep = ';')
df.head()
```

```python
df_year_montant = pd.DataFrame(df.groupby('au_titre_de_l_annee')['montant'].sum())
plt.figure(figsize=(8,8))
explode = [0.05 if year == 2022 else 0 for year in df_year_montant.index]
plt.pie(df_year_montant['montant'], labels = df_year_montant.index, autopct='%1.1f%%', startangle=90, explode = explode)
plt.legend(title='Année', loc='upper left', bbox_to_anchor=(1, 0.5))
plt.title('Subventions accordées entre 2018 et 2022')
plt.legend()
plt.show()
```

```python
df_date = pd.DataFrame(df[['datesperiodeversement', 'montant']].sort_values(by = 'datesperiodeversement').groupby('datesperiodeversement')['montant'].sum().reset_index())
dates_importantes = df_date.sort_values(by = 'montant', ascending = False)['datesperiodeversement'][0:5].to_list()
plt.figure(figsize = (12,8))
sns.set(style="whitegrid")
sns.lineplot(x='datesperiodeversement', y='montant', data=df_date)
plt.xticks(dates_importantes, rotation=45)
plt.xlabel('Date')
plt.ylabel('Montant')
for date_max, montant_max in list(df_date.sort_values(by = 'montant', ascending = False)[0:5].values):
    plt.annotate(f'Montant: {montant_max}',
             xy=(date_max, montant_max), xycoords='data',
             xytext=(20, 20), textcoords='offset points',
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
plt.title('Montant des subventions au cours du temps')
plt.legend()
plt.show()
```
